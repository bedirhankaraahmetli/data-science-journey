## Basic Graph

### Loading Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize graph (dpi specifies pixels per inch. When saving use min 300)
plt.figure(figsize=(5,3), dpi=150)

# Line 1

# Keyword Argument Notation
# plt.plot(x, y, label='2x', color='red', marker='o', linewidth=2, markersize=6, markeredgecolor='blue', linestyle='--')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'ro--', label='2x')

## Line 2

# Selecting interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'b', label='X^2')

# Plot remainder of graph as a dashes
plt.plot(x2[5:], x2[5:]**2, 'b--')

# Add a title 
plt.title('Basic Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 12})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmakrs (scale of graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Saving a figure
plt.savefig('basic_graph.png', dpi=300)

# Show plot
plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())
    
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

# Real World Examples

## Line Graph

```python
gas = pd.read_csv('./data/gas_prices.csv')
```

```python
gas.head()
```

```python
plt.figure(figsize=(10,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 16})

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')

plt.xticks(gas.Year[::2])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('gas_price_figure.png', dpi=300)

plt.show()
```

## Histograms

### Fifa Data

```python
fifa = pd.read_csv('./data/fifa_data.csv')
```

```python
fifa.head()
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

## Pie Charts

```python
fifa.head()
```

```python
fifa['Preferred Foot']
```

```python
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
```

```python

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
```

```python
fifa.Weight.head()
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175,200', 'Over 200']
explode = (.4, .05 , 0, 0, 0.4)
```

```python
plt.style.use('ggplot')

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

## Box & Whisker Plot

```python
fifa.head()

import warnings

warnings.simplefilter(action='ignore', category=DeprecationWarning)
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

labels = ['FC Barcelona', 'Real Madrid']

boxes = plt.boxplot([barcelona, madrid], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)

    # Change Fill Color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.savefig('box_and_whisker.png', dpi=300)
plt.show()
```
